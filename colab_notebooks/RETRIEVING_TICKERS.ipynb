Φτιαχνουμε νεες στηλες για ημερομηνιες (καθημερινες) που ηταν την επομενη, σε 90 , σε 180 και σε 360 ημερες από την επόμενη του 10 k filing

```python
from datetime import timedelta
def next_weekday(date):
    date += timedelta(days=1)
    while date.weekday() >= 5:  # 5 = Saturday, 6 = Sunday
        date += timedelta(days=1)
    return date
```

```python
df_8['filing_date'] = pd.to_datetime(df_8['filing_date'], dayfirst=True, errors='coerce')
```

```python
df_8['0_days_after_report'] = df_8['filing_date'].apply(lambda x: next_weekday(x + timedelta(days=0)) if pd.notnull(x) else pd.NaT)
df_8['close_0_days_after_report'] = ''
df_8['90_days_after_report'] = df_8['filing_date'].apply(lambda x: next_weekday(x + timedelta(days=90)) if pd.notnull(x) else pd.NaT)
df_8['close_90_days_after_report'] = ''
df_8['180_days_after_report'] = df_8['filing_date'].apply(lambda x: next_weekday(x + timedelta(days=180)) if pd.notnull(x) else pd.NaT)
df_8['close_180_days_after_report'] = ''
df_8['360_days_after_report'] = df_8['filing_date'].apply(lambda x: next_weekday(x + timedelta(days=360)) if pd.notnull(x) else pd.NaT)
df_8['close_360_days_after_report'] = ''
```
